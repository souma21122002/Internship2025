```python
import cv2
import numpy as np
import random
import pickle
import os




# Function to apply SRM filters using loaded weights
def apply_srm_filters(image, srm_weights):
    steps = srm_weights.shape[-1]  # Number of filters
    filtered_channels = []

    for step in range(steps):
        filtered_channels_step = []
        for channel in range(3):  # Apply each filter to each color channel
            filter_kernel = srm_weights[:, :, 0, step]
            filtered_image = cv2.filter2D(image[:, :, channel], -1, filter_kernel)
            filtered_channels_step.append(filtered_image)

        combined_residual = np.stack(filtered_channels_step, axis=-1)
        filtered_channels.append(combined_residual)

    combined_residual = np.mean(filtered_channels, axis=0)  # Average over all filters
    combined_residual = np.clip(combined_residual, -3, 3)  # Clipping

    # Normalize to the range [0, 255]
    combined_residual = ((combined_residual - combined_residual.min()) /
                         (combined_residual.max() - combined_residual.min())) * 255.0

    return combined_residual.astype(np.uint8)

# Prepare image for SRM using loaded weights (apply SRM before resizing)
def prepare_srm_image(image_path, srm_weights, image_size=(128, 128)):
    # Check if the image file exists
    if not os.path.exists(image_path):
        print(f"Warning: Image file not found at {image_path}")
        # Return a zero array for the image in case of file not found
        return np.zeros(image_size[0] * image_size[1] * 3) / 255.0 # Scale to 0-1 range

    image = cv2.imread(image_path, cv2.IMREAD_COLOR)

    # Check if image was loaded successfully
    if image is None:
        print(f"Warning: Could not load image from {image_path}")
        # Return a zero array for the image in case of loading failure
        return np.zeros(image_size[0] * image_size[1] * 3) / 255.0 # Scale to 0-1 range

    # Apply SRM filters before resizing
    combined_residual = apply_srm_filters(image, srm_weights)

    # Resize the SRM filtered image
    resized_residual = cv2.resize(combined_residual, image_size)

    return resized_residual.flatten() / 255.0

# Preprocess images with SRM only
def preprocess_images(base_path, label, srm_weights, image_size=(128, 128), max_images=None):
    # Remove X_ela as we are not doing ELA
    X_srm = []
    Y = []
    file_log = []

    # Check if the base directory exists
    if not os.path.isdir(base_path):
        print(f"Warning: Directory not found at {base_path}")
        return [], [], [], [] # Return empty lists if directory doesn't exist


    for dirname, _, filenames in os.walk(base_path):
        filenames.sort()
        for filename in filenames:
            if filename.endswith(('jpg', 'png', 'jpeg','webp')):
                full_path = os.path.join(dirname, filename)
                # Remove ELA processing
                # X_ela.append(prepare_ela_image(full_path, image_size=image_size))

                # Process SRM
                srm_processed = prepare_srm_image(full_path, srm_weights, image_size=image_size)
                # Check if srm_processed is not None (in case of error)
                if srm_processed is not None:
                    X_srm.append(srm_processed)
                    Y.append(label)
                    file_log.append(full_path)

                if len(Y) % 500 == 0:
                    print(f'Processing {len(Y)} images')
                if max_images and len(Y) >= max_images:
                    # Return only SRM data
                    return X_srm, Y, file_log
    # Return only SRM data
    return X_srm, Y, file_log

# Save preprocessed data (only SRM data)
def save_preprocessed_data(X_srm, Y, file_log, filename='CASIAV2data.pkl'):
    # Save only SRM data
    with open(filename, 'wb') as f:
        pickle.dump((X_srm, Y, file_log), f)

def main():
    image_size = (128, 128)

    real_path = '/content/drive/MyDrive/ML_Intership/input_real' # Add your real image path here
    fake_path = '/content/drive/MyDrive/ML_Intership/input_fake' # Add your fake image path here

    # Load SRM weights
    srm_weights = np.load('SRM_Kernels.npy')

    # Normalize each filter by its maximum absolute value
    for i in range(srm_weights.shape[-1]):
        max_value = np.max(np.abs(srm_weights[:, :, 0, i]))
        if max_value != 0:
            srm_weights[:, :, 0, i] /= max_value

    # Process real and fake images with SRM only
    X_real_srm, Y_real, log_real = preprocess_images(real_path, 1, srm_weights, image_size=image_size)
    X_fake_srm, Y_fake, log_fake = preprocess_images(fake_path, 0, srm_weights, image_size=image_size)

    # Combine SRM data
    X_srm = X_real_srm + X_fake_srm
    Y = Y_real + Y_fake
    log = log_real + log_fake

    # Combine all data for shuffling (only SRM data)
    combined = list(zip(X_srm, Y, log))
    random.shuffle(combined)
    # Unpack the shuffled data
    X_srm[:], Y[:], log[:] = zip(*combined)

    print(f'Total images processed: {len(X_srm)}')

    # Save only SRM data
    save_preprocessed_data(X_srm, Y, log, filename='CASIAV2data.pkl')

if __name__ == '__main__':
    main()
```

```python
import cv2
import numpy as np
import random
import pickle
import os
import matplotlib.pyplot as plt

def save_srm_image(processed_image, original_path, output_dir):

    os.makedirs(output_dir, exist_ok=True)


    filename = os.path.splitext(os.path.basename(original_path))[0]

    output_filename = f"{filename}_SRM.jpg"
    output_path = os.path.join(output_dir, output_filename)

    if processed_image.shape[-1] == 3:  # If RGB
        output_image = cv2.cvtColor(processed_image, cv2.COLOR_RGB2BGR)
    else:  # If grayscale
        output_image = processed_image

    # Save the image
    cv2.imwrite(output_path, output_image)
    return output_path

def apply_srm_filters(image, srm_weights):
    steps = srm_weights.shape[-1]  # Number of filters
    filtered_channels = []

    for step in range(steps):
        filtered_channels_step = []
        for channel in range(3):  # Apply each filter to each color channel
            filter_kernel = srm_weights[:, :, 0, step]
            filtered_image = cv2.filter2D(image[:, :, channel], -1, filter_kernel)
            filtered_channels_step.append(filtered_image)

        combined_residual = np.stack(filtered_channels_step, axis=-1)
        filtered_channels.append(combined_residual)

    combined_residual = np.mean(filtered_channels, axis=0)  # Average over all filters
    combined_residual = np.clip(combined_residual, -3, 3)  # Clipping

    # Normalize to the range [0, 255]
    combined_residual = ((combined_residual - combined_residual.min()) /
                         (combined_residual.max() - combined_residual.min())) * 255.0

    return combined_residual.astype(np.uint8)

def prepare_srm_image(image_path, srm_weights, image_size=(128, 128), output_dir=None):
    # Check if the image file exists
    '''if not os.path.exists(image_path):
        print(f"Warning: Image file not found at {image_path}")
        return None, np.zeros(image_size[0] * image_size[1] * 3) / 255.0'''

    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    original_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) if image is not None else None

    if image is None:
        print(f"Warning: Could not load image from {image_path}")
        return None, np.zeros(image_size[0] * image_size[1] * 3) / 255.0

    # Apply SRM filters
    combined_residual = apply_srm_filters(image, srm_weights)

    # Save the output image if output directory is provided
    if output_dir is not None:
        save_srm_image(combined_residual, image_path, output_dir)

    # Resize the SRM filtered image
    resized_residual = cv2.resize(combined_residual, image_size)

    return original_image, resized_residual.flatten() / 255.0

def preprocess_images(base_path, label, srm_weights, image_size=(128, 128), max_images=None, output_dir=None):
    X_srm = []
    Y = []
    file_log = []
    original_images = []

    if not os.path.isdir(base_path):
        print(f"Warning: Directory not found at {base_path}")
        return [], [], [], []

    for dirname, _, filenames in os.walk(base_path):
        filenames.sort()
        for filename in filenames:
            if filename.endswith(('jpg', 'png', 'jpeg','webp')):
                full_path = os.path.join(dirname, filename)

                original_img, srm_processed = prepare_srm_image(
                    full_path, srm_weights, image_size=image_size, output_dir=output_dir
                )

                if srm_processed is not None:
                    original_images.append(original_img)
                    X_srm.append(srm_processed)
                    Y.append(label)
                    file_log.append(full_path)

                if len(Y) % 500 == 0:
                    print(f'Processing {len(Y)} images')
                if max_images and len(Y) >= max_images:
                    return original_images, X_srm, Y, file_log
    return original_images, X_srm, Y, file_log

def main():
    image_size = (128, 128)

    real_path = '/content/drive/MyDrive/ML_Intership/input_real'
    fake_path = '/content/drive/MyDrive/ML_Intership/input_fake'

    # Output directories for processed images
    real_output_dir = '/content/drive/MyDrive/ML_Intership/input_SRM_processed'
    fake_output_dir = '/content/drive/MyDrive/ML_Intership/output_SRM_processed'

    srm_weights = np.load('SRM_Kernels.npy')

    # Normalize each filter
    for i in range(srm_weights.shape[-1]):
        max_value = np.max(np.abs(srm_weights[:, :, 0, i]))
        if max_value != 0:
            srm_weights[:, :, 0, i] /= max_value

    # Process images and save outputs
    orig_real, X_real_srm, Y_real, log_real = preprocess_images(
        real_path, 1, srm_weights, image_size=image_size, output_dir=real_output_dir
    )
    orig_fake, X_fake_srm, Y_fake, log_fake = preprocess_images(
        fake_path, 0, srm_weights, image_size=image_size, output_dir=fake_output_dir
    )

    # Combine and shuffle data
    X_srm = X_real_srm + X_fake_srm
    Y = Y_real + Y_fake
    log = log_real + log_fake

    combined = list(zip(X_srm, Y, log))
    random.shuffle(combined)
    X_srm[:], Y[:], log[:] = zip(*combined)

    print(f'Total images processed: {len(X_srm)}')

    # Save preprocessed data
    #save_preprocessed_data(X_srm, Y, log, filename='CASIAV2data.pkl')

    # Visualization
    num_samples = min(5, len(orig_real), len(orig_fake))
    plt.figure(figsize=(15, 10))

    for i in range(num_samples):
        # Original real image
        plt.subplot(4, num_samples, i+1)
        plt.imshow(orig_real[i])
        plt.title(f'Real Original {i+1}')
        plt.axis('off')

        # SRM processed real image
        plt.subplot(4, num_samples, num_samples+i+1)
        processed_img = (X_real_srm[i].reshape(image_size[0], image_size[1], 3) * 255).astype(np.uint8)
        plt.imshow(processed_img)
        plt.title(f'Real SRM {i+1}')
        plt.axis('off')

    for i in range(num_samples):
        # Original fake image
        plt.subplot(4, num_samples, 2*num_samples+i+1)
        plt.imshow(orig_fake[i])
        plt.title(f'Fake Original {i+1}')
        plt.axis('off')

        # SRM processed fake image
        plt.subplot(4, num_samples, 3*num_samples+i+1)
        processed_img = (X_fake_srm[i].reshape(image_size[0], image_size[1], 3) * 255).astype(np.uint8)
        plt.imshow(processed_img)
        plt.title(f'Fake SRM {i+1}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    main()
```
